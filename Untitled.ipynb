```python
import pandas as pd

# This tells Python to open your spreadsheet file
df = pd.read_csv("USDA.csv")

# These are the 6 nutritional pillars we chose earlier
selected_columns = ['Protein', 'TotalFat', 'SaturatedFat', 'Carbohydrate', 'Sugar', 'Sodium']

# This tells Python to display the first 5 rows of those columns
df[selected_columns].head()
```

```python
# Check for missing values in our 6 columns
df[selected_columns].isnull().sum()
```

```python
# 1. Calculate the averages and fill the missing values
df['SaturatedFat'] = df['SaturatedFat'].fillna(df['SaturatedFat'].mean())
df['Sugar'] = df['Sugar'].fillna(df['Sugar'].mean())

# 2. Let's double check to make sure the missing values are now 0
df[selected_columns].isnull().sum()
```

```python
# Loop through all 6 columns and fill any missing spaces with the average of that column
for col in selected_columns:
    df[col] = df[col].fillna(df[col].mean())

# Double check one last time to make sure EVERYTHING is 0
df[selected_columns].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Create the scaler tool
scaler = StandardScaler()

# 2. Scale our 6 nutrition columns so they are on the same math scale
X_scaled = scaler.fit_transform(df[selected_columns])

# 3. Look at what the numbers look like now for the first row
X_scaled[0]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Test cluster sizes from 1 to 10
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# 2. Plot the "Elbow" graph
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('The Elbow Method to Find Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Error Rate)')
plt.show()
```

```python
# 1. Initialize the KMeans algorithm with 4 groups
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)

# 2. Run the math and assign a cluster label (0, 1, 2, or 3) to each food
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 3. Count how many food items fell into each of the 4 groups
df['Cluster'].value_counts()
```

```python
# Calculate the average nutrients for each cluster
cluster_averages = df.groupby('Cluster')[selected_columns].mean()

# Round the numbers so they are easy to read and display the table
cluster_averages.round(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and size of the plot
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Create the scatter plot using Carbohydrates and Total Fat as the axes
# We use 'hue' to color the dots based on the Cluster labels the AI assigned
sns.scatterplot(
    data=df, 
    x='Carbohydrate', 
    y='TotalFat', 
    hue='Cluster', 
    palette='viridis', 
    alpha=0.6, 
    edgecolor=None
)

# Add titles and labels
plt.title('Nutritional Phenotypes: K-Means Clustering (K=4)', fontsize=14, fontweight='bold')
plt.xlabel('Carbohydrates (g per 100g)', fontsize=12)
plt.ylabel('Total Fat (g per 100g)', fontsize=12)
plt.legend(title='Cluster ID', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout and display the graph
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean value of every nutrient for each of the 4 clusters
cluster_profiles = df.groupby('Cluster')[['Protein', 'TotalFat', 'SaturatedFat', 'Carbohydrate', 'Sugar', 'Sodium']].mean()

# Round the numbers to 2 decimal places to make it look clean and professional
cluster_profiles = cluster_profiles.round(2)

# Display the profile table
cluster_profiles
```
